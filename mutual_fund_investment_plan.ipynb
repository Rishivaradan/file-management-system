```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# User Inputs
initial_investment = float(input("Enter the initial lump sum investment amount ($): "))
monthly_contribution = float(input("Enter the monthly contribution amount ($): "))
annual_return = float(input("Enter the expected annual return (as a decimal, e.g., 0.07 for 7%): "))
investment_duration = int(input("Enter the investment duration (in years): "))
```

```python
# Convert annual return to monthly return
monthly_return = (1 + annual_return) ** (1/12) - 1
```

```python
# Number of months to invest
months = investment_duration * 12
```

```python
# Initialize investment growth array
investment_growth = np.zeros(months)
investment_value = initial_investment
```

```python
# Simulate investment growth
for i in range(months):
    # Apply monthly return
    investment_value = investment_value * (1 + monthly_return) + monthly_contribution
    investment_growth[i] = investment_value
```

```python
# Create a DataFrame for better visualization
dates = pd.date_range(start=pd.Timestamp.today(), periods=months, freq='M')
investment_df = pd.DataFrame({'Investment Value': investment_growth}, index=dates)
```

```python
# Plot the investment growth over time
plt.figure(figsize=(10, 6))
plt.plot(investment_df.index, investment_df['Investment Value'], label='Investment Value')
plt.title('Mutual Fund Investment Growth Over Time')
plt.xlabel('Date')
plt.ylabel('Investment Value ($)')
plt.grid(True)
plt.legend()
plt.show()

# Final investment value
final_value = investment_value
print(f"Final investment value after {investment_duration} years: ${final_value:,.2f}")
```
