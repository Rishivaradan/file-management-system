```python
class Book:
  def __init__(self, title, author, isbn):
    self.title = title
    self.author = author
    self.isbn = isbn
    self.available = True

  def __str__(self):
    return f"Title: {self.title}\nAuthor: {self.author}\nISBN: {self.isbn}\nAvailable: {self.available}"

class Library:
  def __init__(self):
    self.books = []

  def add_book(self, book):
    self.books.append(book)
    print("Book added successfully!")

  def find_book(self, search_term, search_by="title"):
    result = []
    if search_by == "title":
      result = [book for book in self.books if search_term.lower() in book.title.lower()]
    elif search_by == "author":
      result = [book for book in self.books if search_term.lower() in book.author.lower()]
    elif search_by == "isbn":
      result = [book for book in self.books if search_term == book.isbn]
    return result

  def borrow_book(self, isbn):
    for book in self.books:
      if book.isbn == isbn:
        if book.available:
          book.available = False
          print("Book borrowed successfully!")
          return
        else:
          print("Book is currently unavailable.")
          return
    print("Book not found.")

  def return_book(self, isbn):
    for book in self.books:
      if book.isbn == isbn:
        if not book.available:
          book.available = True
          print("Book returned successfully!")
          return
        else:
          print("This book was not borrowed from this library.")
          return
    print("Book not found.")

  def display_all_books(self):
    if not self.books:
      print("The library is currently empty.")
    else:
      for book in self.books:
        print(book, "\n")

library = Library()

book1 = Book("The python", "GUIDO VAN ROSSUM", "BID-0345391803")
book2 = Book("java", "RITCHIE", "BID-0141439518")
book3 = Book("c++", "RITCHIE ", "BID-0061120084")

library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

search_results = library.find_book("pride", search_by="title")
for book in search_results:
  print(book)

library.borrow_book("BID-0345391803")
library.return_book("BID-0345391803")

library.display_all_books()
```
